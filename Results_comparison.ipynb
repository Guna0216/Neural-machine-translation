```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
```

```python
attn_metrics = pd.read_csv('saved_files/attention_metrics.csv')
bi_metrics = pd.read_csv('saved_files/s2s_metrics.csv')
s2s_metrics = pd.read_csv('saved_files/bidirectional_metrics.csv')
```

```python
plt.figure()
plt.plot(attn_metrics["val_loss"], label="Attention")
plt.plot(bi_metrics["val_loss"], label="Bidirectional")
plt.plot(s2s_metrics["val_loss"], label="Seq2Seq")
plt.legend()
plt.ylabel("Validation Loss")
plt.xlabel("Epoch")
plt.show()
```

```python
attn_bleu = pd.read_csv('saved_files/attention_bleu.csv')
bi_bleu = pd.read_csv('saved_files/bidirectional_bleu.csv')
s2s_bleu = pd.read_csv('saved_files/s2s_bleu.csv')
```

```python
attn_bleu = attn_bleu.sort_values('length')
bi_bleu = bi_bleu.sort_values('length')
s2s_bleu = s2s_bleu.sort_values('length')
```

```python
plt.figure()
plt.plot(attn_bleu["length"], attn_bleu["bleu_score"], label="Attention" )
plt.plot(bi_bleu["length"], bi_bleu["bleu_score"], label="Bidirectional")
plt.plot(s2s_bleu["length"], s2s_bleu["bleu_score"], label="Seq2Seq")
plt.legend()
plt.ylabel("Bleu Score")
plt.xlabel("Sentence Length")
plt.show()
```
